```python
import pandas as pd

#### Load data ----
df_total = pd.read_csv('/content/Total.csv')
# create new df with only necessary columns
df_total_time_series = df_total[['PeriodDay', 'Count']].copy()
df_total_time_series['PeriodDay'] = pd.to_datetime(df_total_time_series['PeriodDay'])
df_total_time_series.index = pd.to_datetime(df_total_time_series.index)
#dataframe to be used for time series forecasting
df_total_time_series = df_total_time_series.set_index('PeriodDay')

#set frequency to daily
df_total_time_series = df_total_time_series.asfreq('D')
```

```python
#### Setup experiment ----
from pycaret.time_series import TSForecastingExperiment
exp = TSForecastingExperiment()
exp.setup(data=df_total_time_series, fh=30, session_id=42, numeric_imputation_target='mean')
```

```python
#### Initial Training ----
model = exp.create_model("prophet")
exp.plot_model(model)
```

```python
#### Tune Training ----
tuned_model = exp.tune_model(model)
exp.plot_model(tuned_model)
```

```python
preds_30 = exp.predict_model(tuned_model, fh = 30)
```

```python
final_prediction_for_the_next_month = preds_30['y_pred'].sum()
```

```python
predicted_sales = final_prediction_for_the_next_month.round(0)
```
